### Aim
Write a python program to show Back Propagation Network for the XOR function with Binary Input and Output

```python
import numpy as np
```

```python
# Sigmoid and its derivative
sigmoid = lambda x: 1 / (1 + np.exp(-x))
sigmoid_derivative = lambda x: x * (1 - x)
```

```python
# XOR Input and Output
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([[0], [1], [1], [0]])
```

```python
# Initialize weights
np.random.seed(1)
weights_input_hidden = np.random.uniform(-1, 1, (2, 4))
weights_hidden_output = np.random.uniform(-1, 1, (4, 1))
```

```python
# Training
learning_rate = 0.5
epochs = 10000
```

```python
for i in range(epochs):
    # Forward propagation
    hidden_output = sigmoid(np.dot(X, weights_input_hidden))
    predicted_output = sigmoid(np.dot(hidden_output, weights_hidden_output))

    # Backpropagation
    error = y - predicted_output
    output_delta = error * sigmoid_derivative(predicted_output)
    hidden_delta = output_delta.dot(weights_hidden_output.T) * sigmoid_derivative(hidden_output)

    # Update weights
    weights_hidden_output += hidden_output.T.dot(output_delta) * learning_rate
    weights_input_hidden += X.T.dot(hidden_delta) * learning_rate

# Apply threshold to approximate to 0 or 1
predicted_output_binary = np.round(predicted_output)

# Final output
print("\nPredictions for XOR:")
for i in range(len(X)):
    print(f"Input: {X[i]} -> Predicted: {predicted_output_binary[i]} -> Expected: {y[i]}")
```
