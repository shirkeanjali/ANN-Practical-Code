```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-10, 10, 100)
```

```python
def step(x):
    return np.where(x >= 0, 1, 0)
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def bipolar_sigmoid(x):
    return np.tanh(x)
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def leaky_relu(x):
    return np.where(x > 0, x, 0.01 * x)
```

```python
def elu(x, alpha=1):
    return np.where(x > 0, x, alpha * (np.exp(x) - 1))
```

```python
def softmax(x):
    e_x = np.exp(x - np.max(x))  # stability
    return e_x / e_x.sum()
```

```python
plt.figure(figsize=(15, 10))
```

```python
plt.subplot(3, 3, 1)
plt.plot(x, step(x))
plt.title("Step Function")
```

```python
plt.subplot(3, 3, 2)
plt.plot(x, sigmoid(x))
plt.title("Sigmoid")
```

```python
plt.subplot(3, 3, 3)
plt.plot(x, bipolar_sigmoid(x))
plt.title("Tanh (Bipolar Sigmoid)")
```

```python
plt.subplot(3, 3, 4)
plt.plot(x, relu(x))
plt.title("ReLU")
```

```python
plt.subplot(3, 3, 5)
plt.plot(x, leaky_relu(x))
plt.title("Leaky ReLU")
```

```python
plt.subplot(3, 3, 6)
plt.plot(x, elu(x))
plt.title("ELU")
```

```python
sample = np.array([2.0, 1.0, 0.1])
softmax_output = softmax(sample)

plt.subplot(3, 3, 7)
plt.bar(range(len(sample)), softmax_output)
plt.title("Softmax (Example)")

plt.tight_layout()
plt.show()
```
