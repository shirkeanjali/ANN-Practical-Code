```python
# Step function (activation)
def step_function(y_in, theta):
    if y_in >= theta:
        return 1
    else:
        return 0

# Inputs (all combinations)
inputs = [(0,0), (0,1), (1,0), (1,1)]

# Weights and threshold
w1 = 1     # excitatory
w2 = -1    # inhibitory
theta = 1

print("X1  X2  Y")

# Compute output
for x1, x2 in inputs:
    y_in = x1*w1 + x2*w2
    y = step_function(y_in, theta)
    print(x1, " ", x2, " ", y)
```
